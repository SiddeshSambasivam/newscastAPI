```python
import pandas as pd
from src.utils import parse_data, convert_str_to_datetime, parse_results
import datetime
```

```python
def datetime_to_str(x: str) -> str:
    '''converts datetime object to string'''
    return x.strftime("%d/%m/%Y, %H:%M:%S")
```

```python
df = pd.read_csv('./dumps.csv')
len(df)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df["timestamp"] = df['timestamp'].apply(lambda x: convert_str_to_datetime(x))
df.sort_values(by=["timestamp"], ascending=False, inplace=True)
```

```python
datetime.datetime(2021, 1, 20)
```

```python
dev = df[(df["timestamp"]>= datetime.datetime(2021, 1, 15)) & (df["timestamp"]<= datetime.datetime(2021, 1, 18, 23, 59,59))]
dev.reset_index(drop=True, inplace=True)
dev['timestamp'] = dev['timestamp'].apply(lambda x: datetime_to_str(x)) 
dev
```
